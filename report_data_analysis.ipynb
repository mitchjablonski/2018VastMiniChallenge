```python
def plot_data(my_data):
    interval = alt.selection_interval(encodings=['x'])
    chart1 = alt.Chart(my_data).mark_point().encode(
        alt.X('month_day_yr:T', timeUnit='yearmonthdatehours', axis=alt.Axis(title='Month Date Year')),
        y='Etype:N',
        color=alt.condition(interval,'Etype:N',alt.value('lightgray')),
        tooltip=['Source_Names','Destination_Names', 'month_day_yr:T']
    ).interactive().properties(selection=interval)

    chart2 = alt.Chart(my_data).mark_point().encode(
        alt.X('month_day_yr:T', timeUnit='yearmonthdatehours', axis=alt.Axis(title='Month Date Year')),
        y='Etype:N',
        color=alt.condition(interval,'Etype:N',alt.value('lightgray')),
        tooltip=['Source_Names','Destination_Names', 'month_day_yr:T']
    ).interactive().transform_filter(interval)


    hist = alt.Chart(my_data).mark_bar().encode(
        y='Etype:N',
        x='count()',
        color='Etype:N'
    ).transform_filter(
        interval
    )

    return (chart1 | chart2) & hist
```

```python
my_data = pd.read_csv(files[0])
my_data['Etype'].replace({0:'calls',1:'emails',2:'purchases',3:'meetings',5:'original purchase'}, inplace=True)
my_data.head()
```

```python
plots = plot_data(my_data)
plots
```
