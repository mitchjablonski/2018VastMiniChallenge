```python
data_des = pd.read_csv('data_described_frequency_v2.csv')
```

```python
data_des.head()
```

```python
alt.Chart(data_des).mark_bar().encode(
    x='date_time:T',
    y='change_in_unique_ids',
    color='data_type',
    tooltip='date_time:T'
)
```

```python
interval = alt.selection_interval(encodings=['x','y'])
```

```python
ch1 = alt.Chart(data_des).mark_point().encode(
    x=alt.X('date_time:T', timeUnit='yearmonthdatehours', axis=alt.Axis(title='Transformed dt')),
    y='data_amount',
    color='data_type'
).properties(selection=interval)
```

```python
ch2 = alt.Chart(data_des).mark_bar().encode(
    x='sum(data_amount)',
    y='data_type',
    color='data_type'
).transform_filter(interval)
```

```python
ch3 = alt.Chart(data_des,
         width=200,
    height=200).mark_point().encode(
    x='data_type',
    y='sum(data_amount)',
    color='data_type',
    size='sum(data_amount)',
    tooltip=['sum(data_amount)']
).transform_filter(interval)
```

```python
ch1 | (ch2 & ch3)
```

```python
conf_sus = pd.read_csv('purchase_communication_results/confirmed_suspicious_group_structure_for_1505882733_Rosalia Larroque_Jenice Savaria_where_meeting_found_first_pass.csv')
```

```python
conf_sus.head()
```

```python

'''
0 is for calls
1 is for emails
2 is for purchases
3 is for meetings
'''
conf_sus['Etype'].replace({0:'calls', 1:'emails', 1:'purchases', 3:'meetings'}, inplace=True)
```

```python
alt.Chart(conf_sus).mark_point().encode(
    x=alt.X('full_date:T', timeUnit='yearmonthdatehours', axis=alt.Axis(title='Transformed dt')),
    y='Etype',
    color='Etype',
    tooltip=['Source_Names', 'Destination_Names','full_date']
)
```

```python
interval = alt.selection_interval(encodings=['x'])
```

```python
ch2 = alt.Chart(conf_sus).mark_point().encode(
    x=alt.X('full_date:T', timeUnit='yearmonthdatehours', axis=alt.Axis(title='Transformed dt')),
    y='Etype',
    color=alt.condition(interval,'Etype',alt.value('lightgray')),
    tooltip=['Source_Names', 'Destination_Names','full_date']
).properties(selection=interval)
```

```python
ch1 = alt.Chart(conf_sus).mark_bar().encode(
    x='count()',
    y='Etype',
    color='Etype'
).transform_filter(interval)
```

```python
ch3 = (ch2 & ch1)
```

```python
ch3.save('confirmed_suspicious_over_time.html')
```
