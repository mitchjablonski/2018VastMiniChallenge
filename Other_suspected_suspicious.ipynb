```python
import altair as alt
import pandas as pd
```

```python
sus_purc = pd.read_csv('suspected_suspicious/suspected_suspicous_437025_2037766_1496425862.csv')
```

```python
sus_purc.head()
```

```python
sus_purc.columns
```

```python
alt.Chart(sus_purc).mark_point().encode(
    x='month_day_yr:T',
    y='Etype:N',
    color='Etype:N',
    tooltip=['Source_Names','Destination_Names']
).interactive()
```
