```python
import altair as alt
import pandas as pd

##Interesting observation, the company seems to plumet in terms of business performed ~ 5 months before suspicous data
##begins
```

```python
my_data = pd.read_csv('data_described_frequency_v2.csv')
```

```python
my_data.head()
```

```python
call_data = my_data[my_data['data_type'] == 'calls']
alt.Chart(call_data).mark_point().encode(
    x ='date_time:T',
    y='avg_time_between_interactions',
    color='data_type'
)
```

```python
mtg_data = my_data[my_data['data_type'] == 'meetings']
alt.Chart(mtg_data).mark_point().encode(
    x ='date_time:T',
    y='avg_time_between_interactions',
    color='data_type'
).interactive()
```

```python
email_data = my_data[my_data['data_type'] == 'emails']
alt.Chart(email_data).mark_point().encode(
    x ='date_time:T',
    y='avg_time_between_interactions',
    color='data_type'
)
```

```python
purchase_data = my_data[my_data['data_type'] == 'purchases']
alt.Chart(purchase_data).mark_point().encode(
    x ='date_time:T',
    y='avg_time_between_interactions',
    color='data_type'
)
```

```python
all_but_mtg = my_data[my_data['data_type'] != 'meetings']
alt.Chart(all_but_mtg).mark_point().encode(
    x ='date_time:T',
    y='avg_time_between_interactions',
    color='data_type'
)
```
